##Visualise PCs

```python
def plot(X,pcs):
    d,n =X.shape
    plt.scatter(X[0,:],X[1,:])
    plt.axhline(color='black', linestyle="--")
    plt.axvline(color='black', linestyle="--")
    plt.ylim(np.min(X[0,:])-0.5, np.max(X[0:,])+0.5)
    plt.xlim(np.min(X[0,:])-0.5, np.max(X[0:,])+0.5)
    plt.axis('equal')

    #PC 1
    m_1=pcs[:,0][1]/ pcs[:,0][0] #W_12/w_11

    #pc2
    m_2=pcs[:,1][1]/ pcs[:,1][0] #W_22/w_21

    x=np.linspace(np.min(X[0,:]), np.max(X[0,:]))
    plt.plot(x,m_1*x, color="green", label="PC-1")
    plt.plot(x, m_2*x, color="red", label="PC-2")
    plt.legend()
```

#PCA Function

```python
def PCA(X):
    #shape
    d,n=X.shape
    #center
    X-=X.mean(axis=1).reshape(d,1)
    #Cov Mat
    C=X@X.T/n
    #Eigen decomposition
    eigval, eigvec= np.linalg.eigh(C)
    #arrange in descending
    var=np.flip(eigval)
    pcs=np.flip(eigvec, axis=1)
    return var, pcs
```

```python
mu=np.array([2,5])
cov=np.array([
    [1,-0.9],
    [-0.9,1]
])
n=100
X= rng.multivariate_normal(mu,cov,n).T
var,pcs= PCA(X)
plot(X, pcs)
```

```python
# mus: (2,3)
# mus[:,0] : mean of cluster 1
# mus[:,1] : mean of cluster 2
# mus[:,2] : mean of cluster 3
```

```python
def k_means(X, k=3):
    # X: d x n matrix where each column is a data point in d dimensions
    # k: number of clusters

    d, n = X.shape  # d = dimensions, n = number of data points

    # 🔹 Step 0: Initialize cluster centers randomly
    ind = rng.choice(np.arange(n), k, replace=False)  # randomly select k unique indices
    mus = X[:, ind]  # initial cluster centers (mus) as columns from X

    # 🔹 Initialize cluster assignments
    z = np.ones(n)         # current cluster assignments for each point
    z_prev = np.zeros(n)   # previous assignments to check for convergence

    # 🔁 Repeat until cluster assignments stop changing
    while not np.array_equal(z, z_prev):
        z_prev = z.copy()  # store previous assignments

        # 🔹 Step 1: Assign each point to the nearest cluster center
        for i in range(n):
            # Compute Euclidean distance from point X[:,i] to each cluster center
            dist = np.linalg.norm(mus - X[:, i].reshape(d, 1), axis=0)
            z[i] = np.argmin(dist)  # assign to the closest cluster (minimum distance)

        # 🔹 Step 2: Recompute cluster centers based on current assignments
        for j in range(k):
            if np.any(z == j):  # check if any points are assigned to cluster j
                # Update cluster center j to mean of all points assigned to it
                mus[:, j] = X[:, z == j].mean(axis=1)

    return z  # Final cluster assignments for each data point

# i : index over data points
# j : index over clusters
```
